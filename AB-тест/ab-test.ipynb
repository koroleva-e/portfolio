(—Ç.–µ. —É–¥–∞–ª–∏—Ç—å —Å—Ç—Ä–æ–∫–∏, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ–±—ã—Ç–∏—è –±–æ–ª—å—à–µ 14 –¥–Ω–µ–π)

```python
# –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è 'first_dt'
first_ts = (df
            .sort_values(by=['user_id', 'event_dt'])
            .groupby('user_id')
            .agg({'event_dt': 'first'})
            .rename(columns={'event_dt': 'first_dt'})
            .reset_index()  # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º user_id –∏–∑ –∏–Ω–¥–µ–∫—Å–∞
    )
first_ts
```

```python
# –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–∏–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–≥–æ –≤–∏–∑–∏—Ç–∞ 'first_dt' –∫ 'df'
df = df.merge(first_ts[['user_id', 'first_dt']], on='user_id', how='left')
df
```

```python
# –∑–∞–¥–∞–µ–º –≥–æ—Ä–∏–∑–æ–Ω—Ç –∞–Ω–∞–ª–∏–∑–∞ - 14 –¥–Ω–µ–π
horizon = pd.Timedelta(days=14)

# –ø–æ–¥—Å—á–∏—Ç–∞–µ–º –≤—Ä–µ–º–µ–Ω–Ω–æ–π –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –æ—Ç –ø–µ—Ä–≤–æ–≥–æ –¥–æ —Ç–µ–∫—É—â–µ–≥–æ —Å–æ–±—ã—Ç–∏—è
df['lifetime'] = df['event_date'] - df['first_dt']

# –æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø–æ–ø–∞–¥–∞—é—Ç –≤ –≥–æ—Ä–∏–∑–æ–Ω—Ç
df = df[df['lifetime'] <= horizon].copy()
df
```

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

#### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π

```python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π
events_count = (df.groupby('event_name')
                  .agg({'user_id':'count'})
                  .sort_values(by='user_id', ascending=False)
                  .reset_index()
                  .rename(columns={'user_id' : 'event_count'}))

# –¥–æ–ª—è —Å–æ–±—ã—Ç–∏–π
events_count['event_percent'] = events_count['event_count'] / df['event_name'].count()
events_count.style.format({'event_percent':'{:.2%}'})
```

```python
# —Å—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É 
fig = px.bar(events_count.sort_values(by='event_count', ascending=False),
             x='event_count',
             y='event_name', 
             width=900, height=570,
             text='event_count', 
             color='event_name',
             color_discrete_sequence=px.colors.qualitative.Set2)

fig.update_layout(title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π',
                   xaxis_title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π',
                   yaxis_title='–°–æ–±—ã—Ç–∏—è',
                   showlegend=False)

fig.show() 
```

```python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π –ø–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º
unique_events_count = (df.groupby('event_name')
                         .agg({'user_id': 'nunique'})
                         .reset_index() 
                         .rename(columns={'user_id' : 'unique_user_count'})
                         .sort_values(by='unique_user_count', ascending=False))

# –¥–æ–ª—è —Å–æ–±—ã—Ç–∏–π –ø–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º
unique_events_count['event_percent'] = unique_events_count['unique_user_count'] / unique_events_count['unique_user_count'].sum()
unique_events_count.style.format({'event_percent':'{:.2%}'})
```

- –≤ —Å—Ä–µ–¥–Ω–µ–º, –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≥—Ä—É–ø–ø—ã –ê –¥–µ–ª–∞—é—Ç –Ω–∞ 1 —Å–æ–±—ã—Ç–∏–µ –±–æ–ª—å—à–µ, —á–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≥—Ä—É–ø–ø—ã B

–ü—Ä–æ–≤–µ—Ä–∏–º, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞ —Ä–∞–∑–Ω–∏—Ü–∞ –≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ —Å–æ–±—ã—Ç–∏–π c –ø–æ–º–æ—â—å—é t-—Ç–µ—Å—Ç–∞

**–í—ã–¥–≤–∏–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—ã:**
- H‚ÇÄ: –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –≥—Ä—É–ø–ø–∞—Ö A –∏ B —Ä–∞–≤–Ω–æ
- H‚ÇÅ: –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –≥—Ä—É–ø–ø–∞—Ö A –∏ B —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è

alpha = 0.05

```python
sample_1 = list(events_per_user_group.query('group == "A"')['events_count'])
sample_2 = list(events_per_user_group.query('group == "B"')['events_count'])

alpha = 0.05 # —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –Ω–µ–≥–æ, –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

results = st.ttest_ind(sample_1, sample_2)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É') 
```

```python
# —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ –≤–æ—Ä–æ–Ω–∫–µ (–∞–±—Å–æ–ª—é—Ç–Ω–∞—è –∫–æ–Ω–≤–µ—Ä—Å–∏—è)
funnel = (df
        .groupby('event_name')
        .agg({'user_id': 'nunique'})
        .reindex(['login', 'product_page', 'product_cart', 'purchase']) # —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä—É–µ–º –∏–Ω–¥–µ–∫—Å 'purchase' —á—Ç–æ–±—ã –±—ã–ª –ø–æ—Å–ª–µ–¥–Ω–∏–º
        .reset_index()  
        .rename(columns={'user_id' : 'unique_user_count'}))

# –ø–æ—Å—á–∏—Ç–∞–µ–º –¥–æ–ª—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ—Ö–æ–¥—è—Ç –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–π —à–∞–≥ –≤–æ—Ä–æ–Ω–∫–∏ –≤ % (A ‚Üí B ‚Üí C)
funnel['user_percent'] = funnel['unique_user_count'] / df['user_id'].nunique()
funnel.style.format({'user_percent':'{:.2%}'})
```

```python
# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏–∏ —Ç–µ—Å—Ç–∞
def z_test(group1, group2, event, alpha):

    def successes_trials(group1, group2, event):
    
        successes = np.array([df.query(f'group == "{group1}" & event_name == "{event}"')['user_id'].nunique(),\
                              df.query(f'group == "{group2}" & event_name == "{event}"')['user_id'].nunique()])

        trials = np.array([df.query(f'group == "{group1}"')['user_id'].nunique(),\
                           df.query(f'group == "{group2}"')['user_id'].nunique()])
        return successes, trials
    
    successes, trials = successes_trials(group1, group2, event) 
   
    # –ü—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ:
    p1 = successes[0]/trials[0]
 
    # –ü—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤–æ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø–µ:
    p2 = successes[1]/trials[1]
    
    print(trials[0], trials[1])
 
    # –ü—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ:
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])
 
    # –†–∞–∑–Ω–∏—Ü–∞ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö
    difference = p1 - p2 
 
    # –°—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –≤ —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è—Ö —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))
 
    # –ó–∞–¥–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (—Å—Ä–µ–¥–Ω–µ–µ 0, —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 1)
    distr = stats.norm(0, 1) 
 
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    print(f'–¢–µ—Å—Ç –¥–ª—è –≥—Ä—É–ø–ø {group1} –∏ {group2} –∏ —Å–æ–±—ã—Ç–∏—è {event}')
 
    print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', p_value)
    if p_value < alpha:
        print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –º–µ–∂–¥—É –¥–æ–ª—è–º–∏ –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞')
    else:
        print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π —Å—á–∏—Ç–∞—Ç—å –¥–æ–ª–∏ —Ä–∞–∑–Ω—ã–º–∏') 
```

```python
z_test('A', 'B', 'product_page', alpha)
```
